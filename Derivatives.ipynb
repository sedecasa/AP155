```python
'''this is problem 3.8'''
import numpy as np
import matplotlib.pyplot as plt

# this is (a)
x = 0.004
def func_1(x):
  return (1.0 - np.cos(x))/(x**2)

def first_dev(x):
  return (1.0 - np.cos(x))*(-2.0/(x**3)) + (1/x**2)*(np.sin(x))

def second_dev(x):
  return (6.0 - 4.0*x*np.sin(x) + (x**2 - 6.0)*np.cos(x))/(x**4)


# this is (b)
def central_diff(func, x, dx):
  return (func(x - dx) - 2.0*func(x) + func(x + dx))/(dx**2.0)

dx = np.logspace(-1, -6, 120)

abs_diff_1 = [central_diff(func_1, 0.004, dxi) - second_dev(0.004) for dxi in dx]

plt.loglog(dx, np.abs(abs_diff_1), "o")
plt.xlabel("h")
plt.ylabel("Absolute Error")
plt.title("Erorr of Second Derivative Approximation")
plt.show()


# this is (c)
def func_2(x):
  return (2.0*(np.sin(x/2)**2))/(x**2)

abs_diff_2 = [central_diff(func_2, 0.004, dxi) - second_dev(0.004) for dxi in dx]
plt.loglog(dx, np.abs(abs_diff_2), "o")
plt.xlabel("h")
plt.ylabel("Absolute Error")
plt.title("Erorr of Second Derivative Approximation")
plt.show()


# comparative graph
plt.loglog(dx, np.abs(abs_diff_1), "o", label='If you do not use the trigonometric identity')
plt.loglog(dx, np.abs(abs_diff_2), "o", label='If you use the trigonometric identity')
plt.xlabel("h")
plt.ylabel("Absolute error")
plt.title("Erorr of Second Derivative Approximation")
plt.legend()
plt.show()
```

```python
'''this is problem 3.25'''
# this is (a)
x = np.linspace(0, 2.0*np.pi, 128)

def f_func(x):
  return 2.0 + 5.0*np.sin(x) + 0.1*np.sin(30*x)

def g_func(x):
  return 2.0 + 5.0*np.sin(x)

plt.plot(x, f_func(x), '-', label='f(x)')
plt.plot(x, g_func(x), '-', label='g(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()


# this is (b)
dx = x[1] - x[0]

def f_prime(x):
  return 5.0*np.cos(x) + 3.0*np.cos(30*x)

def g_prime(x):
  return 5.0*np.cos(x)

def f_dapp(f, x, dx):
  return (f(x + dx) - f(x))/dx

plt.plot(x, f_prime(x), '-', label="analytical f'(x)")
plt.plot(x, g_prime(x), '-', label="analytical g'(x)")
plt.plot(x, f_dapp(f_func, x, dx), '-', label='forward diff (h)')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()


# this is (c)
def f_dapp_2(f, x, dx):
  return (f(x + 2*dx) - f(x))/(2*dx)

plt.plot(x, f_prime(x), '-', label="analytical f'(x)")
plt.plot(x, g_prime(x), '-', label="analytical g'(x)")
plt.plot(x, f_dapp(f_func, x, dx), '-', label='forward diff (h)')
plt.plot(x, f_dapp_2(f_func, x, dx), '-', label='forward diff (2h)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Derivative of a Noisy Function")
plt.legend()
plt.show()
```
