```python
#This is problem 2.6
import numpy as np
a = 1
b = 10**8
c = 1

x_plus = (-b + np.sqrt(b**2 - 4*a*c))/(2*a)
x_minus = (-b - np.sqrt(b**2 - 4*a*c))/(2*a)

print(x_minus)
print(x_plus*x_minus)
print(x_plus)
another_x_plus = c/(a*x_minus)
print(another_x_plus)

#as B increases exponentially, the difference between x_plus and another_x_plus increases
```

```python
#This is problem 2.8
import matplotlib.pyplot as plt
x = np.array([0.1*i for i in range(1, 101)])
y_a = (1 - np.cos(x))/(x**2)

fig, ax = plt.subplots()
ax.plot(x, y_a)
plt.show()
```
