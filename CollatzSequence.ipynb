```python
import matplotlib.pyplot as plt

def collatz_sequence(N):
  i = 0
  while N != 1:
    if N % 2 == 0:
      N = N/2
    else:
      N = 3*N + 1
    i += 1
  return i + 1

N_values = list(range(1, 10001))
TN_values = list(collatz_sequence(n) for n in N_values)

plt.scatter(N_values, TN_values, s=10)
plt.xlabel("N")
plt.ylabel("T(N)")
plt.show()
```
